```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (skip metadata rows)
df = pd.read_csv(r'C:\Users\Public\Documents\API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv', skiprows=4)

# Keep only relevant columns
df_2024 = df[['Country Name', 'Country Code', '2024']].dropna()

# Filter only actual countries (3 uppercase letter country codes)
df_2024 = df_2024[df_2024['Country Code'].str.match(r'^[A-Z]{3}$')]

# ---------------------------
# Bar Chart: Top 10 populous countries
# ---------------------------
top10 = df_2024.sort_values(by='2024', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top10['Country Name'], top10['2024'], color='skyblue', edgecolor='black')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Country Name')
plt.ylabel('Population')
plt.title('Bar Chart of Top 10 Most Populous Countries in 2024')
plt.tight_layout()
plt.show()

# ---------------------------
# Histogram: Population distribution
# ---------------------------
plt.figure(figsize=(8, 5))
plt.hist(df_2024['2024'], bins=20, color='green', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.title('Histogram of Population Distribution (2024)')
plt.tight_layout()
plt.show()
```
